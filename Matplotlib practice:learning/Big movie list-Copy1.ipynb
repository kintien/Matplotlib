```python
import pandas as pd 
import math
import matplotlib.pyplot as plt
file =  "/Users/kintienmalik/Documents/AML300/Student Directory/Kintien/bigmovies2.csv"
data=pd.read_csv(file)



data

```

```python
df =data.loc[:,['Movie', 'Total Domestic Gross']]
df.sort_values(['Total Domestic Gross'])
```

```python
df_total_month =data.loc[:,['Movie','Month','Total Domestic Gross']]
df_total_sorted= df_total_month.groupby('Month').sum()
df_total_sorted.sort_values(['Total Domestic Gross'], ascending =False)
```

```python
df_total_month =data.loc[:,['Movie','Month','Oscar nominations']]
df_total_sorted= df_total_month.groupby('Month').sum()
df_total_sorted.sort_values(['Oscar nominations'], ascending =False)
```

```python
df_rating= data[data['MPAA_rating']== 'PG-13']
df_monthly_tdg= df_rating.loc[:,['Month','MPAA_rating','Total Domestic Gross']]
df_rating_sum=df_monthly_tdg.groupby('Month').sum()
df_rating_sum.sort_values(['Total Domestic Gross'], ascending= False)
```

```python
df_monthly_tdg= df_rating.loc[:,['Genre','Total Domestic Gross']]
df_rating_sum=df_monthly_tdg.groupby('Genre').sum()
df_rating_sum.sort_values(['Total Domestic Gross'], ascending= False)
```

```python
df_monthly_tdg= df_rating.loc[:,['Genre','Total Domestic Gross']]
df_rating_sum=df_monthly_tdg.groupby('Genre').mean()
df_rating_sum.sort_values(['Total Domestic Gross'], ascending= False)
```
