```python
import pandas as pd 
import math
import matplotlib.pyplot as plt

drugsfile = "/Users/kintienmalik/Documents/AML300/Student Directory/Kintien/Drugs_w_name.csv"
drugsfile =pd.read_csv(drugsfile)

Grades = drugsfile['Grades'].replace(['A','B','C','D','F'], [4, 3, 2, 1, 0])
list1=[]
list2=[]
marajuana=[]
alcohol=[]
Vape=[]
Acid=[]

x =drugsfile['Drugs used']
for i in range(0,len(x)):
    list1.append(x[i].split())
for z in range (0,len(x)):
    if list1[z] == ['None']:
        list2.append(0)
    else:
        list2.append(len(list1[z]))
    

for z in range (0,len(x)):
    if 'Marijuana' in list1[z]:
        marajuana.append(1)
    else:
        marajuana.append(0) 
        
        

for z in range (0,len(x)):
    if 'Alcohol' in list1[z]:
        alcohol.append(1)
    else:
        alcohol.append(0) 
        

for z in range (0,len(x)):
    if 'Vape' in list1[z]:
        Vape.append(1)
    else:
        Vape.append(0)     
    

for z in range (0,len(x)):
    if 'Acid' in list1[z]:
        Acid.append(1)
    else:
        Acid.append(0)     
print(marajuana)
    
    
drugsfile['Drugs_Used'] = list1
drugsfile['GPA'] = Grades
drugsfile['Number of drugs being used'] = list2
drugsfile['GPA'] = Grades
drugsfile['Marijuana']= marajuana
drugsfile['Alcohol']= alcohol
drugsfile['Vape']= Vape
drugsfile['Acid']= Acid


drugsfile
```

```python
ax=drugsfile.plot(kind='bar', x= 'survey id #', y= 'Number of drugs being used')
plt.rcParams['figure.figsize'] =(20,10)
ax.set_xlabel("id#")
ax.set_ylabel('Number ofdrugs beig used')
plt.show()
print('Show most people do 1 to 2 drugs')
```

```python
ax= plt.gca()
drugsfile.plot(kind= 'bar', x='GPA',y ='Number of drugs being used', ax=ax)
```

```python
ax =drugsfile.groupby(['GPA','Number of drugs being used'])['Drugs_Used'].size().unstack().plot(kind='bar',stacked=True)
ax.set_ylabel('Number of people')
#the less drugs you do the higher your GPA
```

```python
ax =drugsfile.groupby(['Gender','Number of drugs being used'])['Drugs_Used'].size().unstack().plot(kind='bar',stacked=True)
ax.set_ylabel('Number of people')
plt.show()



#this tells me most males and females do 1 or 2 drugs 
#and males do a greater number of drugs than females
```
